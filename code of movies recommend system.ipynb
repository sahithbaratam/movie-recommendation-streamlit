```python
import pandas as pd

# Load datasets
movies = pd.read_csv('archive (4)/tmdb_5000_movies.csv')
credits = pd.read_csv('archive (4)/tmdb_5000_credits.csv')

# Merge datasets on movie ID
df = movies.merge(credits, left_on='id', right_on='movie_id', how='left')

# Create tags column (example combining genres, keywords, overview)
import ast

def combine_features(row):
    features = []

    # Convert stringified lists to Python lists
    for col in ['genres', 'keywords', 'cast', 'crew']:
        if col in row and pd.notnull(row[col]):
            row_list = ast.literal_eval(row[col])
            features += [x['name'] for x in row_list] if isinstance(row_list, list) else []

    # Add overview text
    if 'overview' in row and pd.notnull(row['overview']):
        features.append(row['overview'])
    
    return features

df['tags'] = df.apply(combine_features, axis=1)

# Clean tags
# Clean tags
new_df = df[['id', 'original_title', 'tags']].copy()  # use correct column names
new_df.rename(columns={'original_title':'title'}, inplace=True)  # optional: rename to 'title'

# Preprocessing
new_df['tags'] = new_df['tags'].fillna('')
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())

```

```python
import pandas as pd

movies = pd.read_csv('archive (4)/tmdb_5000_movies.csv')
credits = pd.read_csv('archive (4)/tmdb_5000_credits.csv')
```

```python
# Merge movies with credits
movies = movies.merge(credits, left_on='id', right_on='movie_id')
```

```python
# Extract and combine relevant features
import ast

def combine_features(row):
    # Convert JSON-like strings to lists
    genres = [i['name'] for i in ast.literal_eval(row['genres'])]
    cast = [i['name'] for i in ast.literal_eval(row['cast'])][:3]  # top 3
    crew = [i['name'] for i in ast.literal_eval(row['crew']) if i['job'] == 'Director']
    return ' '.join(genres + cast + crew)

movies['tags'] = movies.apply(combine_features, axis=1)
movies['tags'] = movies['tags'].str.lower()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()
similarity = cosine_similarity(vectors)

# Save
pickle.dump(movies, open('app_streamlit/movie_dict.pkl', 'wb'))
pickle.dump(similarity, open('app_streamlit/similarity.pkl', 'wb'))
```
